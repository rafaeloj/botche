```python
import bs4 as bs
```

```python
import urllib
import jellyfish as jf
import heapq
import pandas as pd
import urllib.request
```

```python
KEYS_TO_SEARCH = [
    'federated+learning',
    'communication',
    'iot',
    'network',
    'system+distribution',
    'urban+mobility'
]
```

```python
str_keys = '+'.join(KEYS_TO_SEARCH)
url = f"http://www.wikicfp.com/cfp/servlet/tool.search?q={str_keys}&year=f"
print(url)
source = urllib.request.urlopen(url).read()
```

```python
soup = bs.BeautifulSoup(source,'lxml')
```

```python
def get_table_rows_papers(soup):
    return soup.find_all(
        'table',
        {
            'cellpadding': '2',
            'cellspacing': '1',
            'align': 'center',
            'width': '100%'
        }
    )[0].find_all('tr')
```

```python
def get_sail_and_full_name(td):
    raw_sail, full_name = td.text.strip().split("\n")
    sail = raw_sail.replace('2024','').strip()
    return sail, full_name

def get_when_location_deadline(td):
    return td.text.strip().split("\n")

def get_link(pear):
    link = pear.find_all('a', href=True)[0]['href']
    print(f'http://www.wikicfp.com{link}')
    source = urllib.request.urlopen(f'http://www.wikicfp.com{link}').read()
    soup = bs.BeautifulSoup(source,'lxml')
    a = soup.find_all('a', target='_newtab')[0]
    return a['href']

def treat_rows(rows):
    real_rows = rows[1:]
    pear_rows = [(real_rows[i], real_rows[i+1]) for i in range(0,len(real_rows)-1,2)]
    events = []
    for pear in pear_rows:
        sail, full_name = get_sail_and_full_name(pear[0])
        when, location, deadline = get_when_location_deadline(pear[1])
        link = get_link(pear[0])
        events.append({
            'sail': sail,
            'full_name': full_name,
            'when': when,
            'location': location,
            'deadline': deadline,
            'link': link,
        })
    return events
```

```python
conferences = treat_rows(
    get_table_rows_papers(soup)
)
conferences
```

```python
def get_qualis_data():
    url = f"https://docs.google.com/spreadsheets/d/e/2PACX-1vTZsntDnttAWGHA8NZRvdvK5A_FgOAQ_tPMzP7UUf-CHwF_3PHMj_TImyXN2Q_Tmcqm2MqVknpHPoT2/pubhtml?gid=0&single=true"
    source = urllib.request.urlopen(url).read()
    qualis_data = bs.BeautifulSoup(source, 'lxml')
    data = []
    for row in qualis_data.find('tbody'):
        raw_sail, raw_full_name, raw_qualis = row.find_all('td')
        data.append({
            'sail': raw_sail.text,
            'full_name': raw_full_name.text,
            'qualis': raw_qualis.text,
        })
    return data
```

```python
conferences_qualis = get_qualis_data()
conferences_qualis
```

```python
def filter_by_qualis(qualis):
    return filter(lambda x: x['qualis'] == qualis, conferences_qualis)
```

```python
pd.DataFrame(filter_by_qualis('A3'))
```

```python
def get_qualis(conference):
    qualis_to_conference = []
    for qualis in conferences_qualis:
        similarity = jf.levenshtein_distance(conference, qualis['full_name'])
        qualis_to_conference.append({
            **qualis,
            'similarity': similarity,
        })
    return heapq.nsmallest(1, qualis_to_conference, lambda x: x['similarity'])[0]
```

```python
get_qualis('Consumer Communications & Networking Conference')
```

## Get qualis of conference

```python
conferences_with_qualis = []
for conference in conferences:
    qualis = get_qualis(conference=conference['full_name'])
    # print(qualis)
    conferences_with_qualis.append({
        **conference,
        'qualis': qualis['qualis'],
        '_real_conference': qualis['full_name'],
        'similarity': qualis['similarity'],
        '_sail': qualis['sail'],
    })
```
