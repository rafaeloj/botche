```python
import pandas as pd
```

```python

# Dados de exemplo
data = {
    "sail": ["SOMET 2025", "EN-IoT", "RSEJ", "Ei/Scopus-AACIP"],
    "full_name": [
        "24th International Conference on Intelligent Software Methodologies Tools and Techniques",
        "1st International Workshop on Energy Neutral and Sustainable IoT Devices and Infrastructure",
        "Renewable and Sustainable Energy: An International Journal",
        "2024 2nd Asia Conference on Algorithms, Computing and Image Processing (AACIP 2024)-EI Compendex"
    ],
    "when": ["Sep 23, 2025 - Sep 26, 2025", "Oct 21, 2024 - Oct 21, 2024", "N/A", "Sep 20, 2024 - Sep 22, 2024"],
    "location": ["KitaKyushu -Japan", "Paris, France", "N/A", "Chengdu, China"],
    "deadline": ["Apr 1, 2025", "Aug 13, 2024", "Aug 17, 2024", "Aug 22, 2024"],
    "qualis": ["B4", "B4", "A3", "A3"],
    "_full_name": [
        "IEEE International Conference on Imaging Systems and Techniques",
        "International Workshop on Agents in Traffic and Transportation",
        "Conference on Human Information Interaction and Retrieval",
        "International Conference on Systems, Signals and Image Processing"
    ],
    "similarity": [36, 47, 36, 54],
    "_sail": ["IEEE IST", "ATT", "CHIIR", "IWSSIP"]
}

# Criando o DataFrame
df = pd.DataFrame(data)

# Padronizar datas e outros campos
def padronizar_datas(data_str):
    if data_str == "N/A":
        return None
    return pd.to_datetime(data_str, errors='coerce')

df['when_start'] = df['when'].apply(lambda x: padronizar_datas(x.split(' - ')[0]) if x != "N/A" else None)
df['when_end'] = df['when'].apply(lambda x: padronizar_datas(x.split(' - ')[1]) if x != "N/A" else None)
df['deadline'] = pd.to_datetime(df['deadline'], errors='coerce')

# Tratar valores ausentes
# df.fillna({"location": "Unknown", "when_start": None, "when_end": None}, inplace=True)

# Renomear colunas para um padrão
df.rename(columns={
    'sail': 'event_code',
    'full_name': 'event_name',
    'location': 'event_location',
    'deadline': 'submission_deadline',
    'qualis': 'qualis_rating',
    '_full_name': 'related_event',
    'similarity': 'name_similarity',
    '_sail': 'related_event_code'
}, inplace=True)


print(df)
```
